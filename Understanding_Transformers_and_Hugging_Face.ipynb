# Introduction

## What is NLP?

Natural Language Processing (NLP) is a branch of Artificial Intelligence (AI) that enables computers to understand, interpret, process, and generate human language. NLP combines computational linguistics, machine learning, and deep learning techniques to help machines interact with human language in a meaningful way.

### Applications of NLP
- Chatbots and Virtual Assistants
- Machine Translation
- Text Summarization
- Sentiment Analysis
- Question Answering Systems
- Speech Recognition

---

## Limitations of RNNs

Recurrent Neural Networks (RNNs) were among the earliest deep learning architectures used for sequential data processing. Although effective for short sequences, they have several limitations:

### Challenges of RNNs
- **Sequential Processing:** Tokens are processed one after another, making training slow.
- **Vanishing Gradient Problem:** Information from earlier tokens can be lost during backpropagation.
- **Limited Long-Term Memory:** Difficulty in capturing dependencies across long text sequences.
- **Low Parallelization:** Sequential computation prevents efficient parallel processing.

---

## Limitations of LSTMs

Long Short-Term Memory (LSTM) networks were introduced to address the vanishing gradient problem of RNNs. While they improved sequence modeling, several challenges still remained.

### Challenges of LSTMs
- **Sequential Computation:** Processing remains token-by-token.
- **Difficult Parallelization:** Training cannot fully utilize modern parallel hardware.
- **High Computational Cost:** Longer sequences require more computation.
- **Slow Training:** Processing large datasets becomes time-consuming.

---

## Why Transformers Were Introduced

Transformers were introduced to overcome the limitations of RNNs and LSTMs. Unlike previous architectures, Transformers process all tokens simultaneously using an attention mechanism.

### Advantages of Transformers
- **Self-Attention Mechanism:** Captures relationships between all words in a sequence.
- **Parallel Processing:** Enables faster training and inference.
- **Better Long-Range Dependency Handling:** Maintains context across long sequences.
- **Scalability:** Can be trained efficiently on large datasets and modern hardware.

Transformers have become the foundation of many state-of-the-art NLP models, including BERT, GPT, T5, and BART.

---

## The Significance of the *Attention Is All You Need* Paper

The Transformer architecture was introduced in the landmark research paper **"Attention Is All You Need"** by Vaswani et al. in 2017.

### Key Contributions
- Eliminated recurrence and convolution from sequence modeling.
- Introduced the **Self-Attention Mechanism** as the core component.
- Enabled highly parallelizable training.
- Improved performance on machine translation and other NLP tasks.
- Established the foundation for modern Large Language Models (LLMs).

The publication of this paper marked a major breakthrough in Natural Language Processing and paved the way for the development of today's advanced AI systems such as ChatGPT, Gemini, and Claude.

# Transformer Architecture Section

The Transformer architecture is a deep learning model designed for sequence-to-sequence tasks. It relies on attention mechanisms to process input data efficiently and capture long-range dependencies between words.

---

## Input Embeddings

Before a Transformer can process text, each word (or token) is converted into a numerical vector representation called an **embedding**. These embeddings capture the semantic meaning of words in a continuous vector space.

### Example

Input Sentence:

```text
I love AI
```

Embedding Representation:

```text
I      → [0.12, 0.54, ...]
love   → [0.78, 0.34, ...]
AI     → [0.91, 0.27, ...]
```

These vectors are then passed to the next stage of the Transformer.

---

## Positional Encoding

Unlike RNNs and LSTMs, Transformers process all tokens simultaneously. As a result, they do not inherently understand the order of words in a sentence.

To provide positional information, **Positional Encoding** is added to the input embeddings.

### Formula

\[
PE(pos,2i)=\sin\left(\frac{pos}{10000^{2i/d_{model}}}\right)
\]

where:

- **pos** = position of the word in the sequence
- **i** = embedding dimension index
- **d_model** = embedding size

Positional encodings allow the model to distinguish between different word positions.

---

## Self-Attention

Self-Attention enables the Transformer to determine which words are most relevant to each other within a sentence.

Each token is transformed into three vectors:

- **Query (Q)**
- **Key (K)**
- **Value (V)**

### Attention Formula

\[
Attention(Q,K,V)=softmax\left(\frac{QK^T}{\sqrt{d_k}}\right)V
\]

where:

- \(QK^T\) measures similarity between words.
- \(d_k\) is the dimension of the key vectors.
- Softmax converts scores into probabilities.

### Example

Sentence:

```text
The animal didn't cross the street because it was tired.
```

The word **"it"** refers to **"animal"**.

Self-Attention helps the model identify this relationship and focus on the correct word when understanding the sentence.

---

## Multi-Head Attention

Instead of using a single attention mechanism, Transformers use multiple attention heads simultaneously.

Each attention head learns different relationships in the text.

### Examples of Learned Relationships

- Grammar
- Context
- Semantics
- Word Dependencies

This allows the model to capture richer information from the input sequence.

---

## Feed Forward Network (FFN)

After the attention layer, each token passes through a fully connected neural network known as the Feed Forward Network.

### Structure

```text
Linear
  ↓
ReLU
  ↓
Linear
```

The FFN introduces non-linearity and helps the model learn complex patterns.

---

## Residual Connections

Residual Connections help preserve information and improve gradient flow during training.

### Formula

\[
Output = Layer(x + Sublayer(x))
\]

### Benefits

- Prevents information loss
- Enables deeper networks
- Improves training stability

---

## Layer Normalization

Layer Normalization is applied after attention and feed-forward layers.

### Purpose

- Stabilizes training
- Speeds up convergence
- Reduces internal covariate shift
- Improves model performance

Layer Normalization is typically used together with Residual Connections.

---

## Encoder Architecture

The Encoder processes the input sequence and generates contextualized representations.

### Encoder Flow

```text
Input
  ↓
Embedding
  ↓
Positional Encoding
  ↓
Multi-Head Attention
  ↓
Add & Norm
  ↓
Feed Forward Network
  ↓
Add & Norm
```

### Encoder Components

1. Input Embeddings
2. Positional Encoding
3. Multi-Head Self-Attention
4. Residual Connections
5. Layer Normalization
6. Feed Forward Network

Multiple encoder layers are stacked together to build a deep Transformer.

---

## Decoder Architecture

The Decoder generates the output sequence one token at a time.

### Decoder Flow

```text
Masked Multi-Head Attention
            ↓
Encoder-Decoder Attention
            ↓
Feed Forward Network
            ↓
Prediction
```

### Decoder Components

1. Masked Self-Attention
2. Encoder-Decoder Attention
3. Feed Forward Network
4. Residual Connections
5. Layer Normalization
6. Output Prediction Layer

The Decoder uses information from both the previously generated tokens and the Encoder output to generate the next word in the sequence.

## Architecture

BART (Bidirectional and Auto-Regressive Transformers) is an **Encoder-Decoder Transformer** architecture.

### Key Characteristics

- Uses a bidirectional encoder similar to BERT.
- Uses an autoregressive decoder similar to GPT.
- Combines the strengths of both architectures.
- Designed for sequence-to-sequence tasks such as summarization, translation, and text generation.

---

## Intended Use Cases

The BART Large CNN model is primarily designed for:

- Text Summarization
- Document Summarization
- News Article Summarization
- Sequence-to-Sequence Learning Tasks
- Text Generation Applications

This model performs particularly well when generating concise summaries from long-form text.

---

## Training Data

The model was fine-tuned on the **CNN/DailyMail** dataset.

### About the Dataset

- Contains thousands of news articles.
- Includes human-written summaries.
- Widely used as a benchmark dataset for summarization tasks.

The model learns to generate meaningful summaries by mapping long news articles to shorter summary texts.

---

## Evaluation Metrics

The performance of BART Large CNN is commonly evaluated using ROUGE metrics.

### ROUGE Metrics

#### ROUGE-1
Measures overlap of individual words (unigrams) between the generated summary and the reference summary.

#### ROUGE-2
Measures overlap of consecutive word pairs (bigrams).

#### ROUGE-L
Measures the longest common subsequence between the generated and reference summaries.

Higher ROUGE scores generally indicate better summarization quality.

---

## Limitations and Biases

Although BART Large CNN performs well, it has several limitations.

### Hallucinations

The model may generate information that is not present in the source text.

### Context Loss

Important details may sometimes be omitted when generating shorter summaries.

### Bias Inherited from Training Data

Since the model is trained on large text datasets, it may reflect biases present in the training data.

### Domain Dependence

Performance may decrease when summarizing content that differs significantly from news articles.

# Summarization Using AutoClasses

````python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

model_name = "facebook/bart-large-cnn"

tokenizer = AutoTokenizer.from_pretrained(model_name)

model = AutoModelForSeq2SeqLM.from_pretrained(model_name)
````

````python
text = """
Artificial Intelligence is transforming industries across the globe.
AI systems are increasingly being used in healthcare,
finance, education and transportation.
"""
````

````python
inputs = tokenizer(
    text,
    return_tensors="pt",
    max_length=1024,
    truncation=True
)

summary_ids = model.generate(
    inputs["input_ids"],
    max_length=50,
    min_length=10,
    num_beams=4
)

summary = tokenizer.decode(
    summary_ids[0],
    skip_special_tokens=True
)

print(summary)
````
